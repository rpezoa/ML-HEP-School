```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Advertising dataset
url = 'https://www.statlearning.com/s/Advertising.csv'
advertising_data = pd.read_csv(url, index_col=0)
```

```python
advertising_data
```

```python
# Display the first few rows of the dataset
print("First few rows of the Advertising dataset:")
print(advertising_data.head())

# Basic statistics of the dataset
print("\nSummary statistics:")
print(advertising_data.describe())
```

```python
# Plot relationships between features and sales
features = ['TV', 'radio', 'newspaper']
plt.figure(figsize=(15, 5))
for i, feature in enumerate(features):
    plt.subplot(1, 3, i + 1)
    plt.scatter(advertising_data[feature], advertising_data['sales'], alpha=0.7)
    plt.title(f'Sales vs {feature}')
    plt.xlabel(feature)
    plt.ylabel('Sales')
    plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# 3D Plot of TV, Radio, and Sales
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Extract relevant data
X_tv = advertising_data['TV']
X_radio = advertising_data['radio']
X_news = advertising_data['newspaper']

# Create the 3D scatter plot
ax.scatter(X_tv, X_radio, X_news, c='b', marker='o', alpha=0.7)
ax.set_title('Advertisement features')
ax.set_xlabel('TV Advertising')
ax.set_ylabel('Radio Advertising')
ax.set_zlabel('Newspaper Advertising')

plt.show()
```

```python

import plotly.express as px


# Create an interactive 3D scatter plot
fig = px.scatter_3d(advertising_data, x='TV',y='radio',z='newspaper',color='sales',
    title='Interactive 3D Plot of Sales vs TV and Radio',
    labels={'TV': 'TV Advertising', 'Radio': 'Radio Advertising', 'Newspaper Advertising': 'Newspaper'},
    width=700,  # Increase width
    height=800 
)

# Show the plot
fig.show()
```
